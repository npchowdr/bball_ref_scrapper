```python
years = list(range(1980,2022))
```

```python
url_start = "https://www.basketball-reference.com/leagues/NBA_{}_per_game.html"
```

```python
import requests

for year in years:
    url = url_start.format(year)
    data = requests.get(url)
    
    with open("playerPerGameStats/{}.html".format(year), "w", encoding="utf8") as f:
        f.write(data.text)
```

```python
dfs = []
for year in years:
    with open("playerPerGameStats/{}.html".format(year), encoding="utf8") as f:
        page = f.read()
        soup = BeautifulSoup(page, "html.parser")
        
        list_decompose = soup.find_all('tr', class_="thead")
        for item in list_decompose:
            item.decompose()
            
#         list_decompose = soup.find_all('tr', class_=None)
#         for item in list_decompose:
#             item.decompose() 
        
        stats_table = soup.find(id="per_game_stats")
        stats = pd.read_html(str(stats_table))[0]
        stats["Year"] = year
        dfs.append(stats)
    
```

```python
player_stats = pd.concat(dfs)
```
